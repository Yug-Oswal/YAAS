```python
pipe = pipeline(model="facebook/bart-large-mnli")
```

```python
print(pipe("Human resource management (HRM or HR) is the strategic and coherent approach to the effective and efficient management of people in a company or organization such that they help their business gain a competitive advantage. It is designed to maximize employee performance in service of an employer's strategic objectives.[1][need quotation to verify] Human resource management is primarily concerned with the management of people within organizations, focusing on policies and systems.", candidate_labels = ["Technical", "Management", "Design", "Humanities", "Arts"]))
```

```python
import re

f = open("Domlist.txt", "r")
lines = f.readlines()

c = 0
l = []
l.append([]) 
for subdom in lines:
  if (subdom == '\n'):
    c += 1
    l.append([])
  else:
    match = re.search(r"\d+\.\s*(.+)", subdom)
    if match:
        phrase = match.group(1)
        l[c].append(phrase)

dic = {0: "Technical", 1: "Managerial and Business", 2: "Creative Arts", 3: "Research", 4: "Humanitarian Sciences"}

for i in range(5):
  print()
  print(f"{dic[i]} Domain: ")
  print("Subdomains: ")
  for j in range(52):
    if (i == 2):
      if (j > 36):
        break
    else:
      print(l[i][j])
```

```python
import matplotlib.pyplot as plt
import numpy as np

gptinpstr = "Machine Learning Engineer: Machine learning (ML) is a field devoted to understanding and building methods that let machines \"learn\" – that is, methods that leverage data to improve computer performance on some set of tasks."

def classify(gptinpstr):
  for i in range(5):
    x = pipe(gptinpstr, candidate_labels = l[i])
    labels = x['labels']
    print("Labels: ", labels)
    scores = x['scores']
    print("Probabilities: ", scores)
    scores = np.array(scores)
    scores = scores * 100
    if (i == 0):
      print("Domain with best suitability: ", labels[np.argmax(scores)])
      print("Suitability: ", np.max(scores), "\b%")
    plt.scatter(labels, scores)
    plt.xticks([labels[np.argmax(scores)]])
    plt.xlabel("Domains")
    plt.ylabel("Probability of doing well in domain (/100)")
    plt.show()

classify(gptinpstr)
```
